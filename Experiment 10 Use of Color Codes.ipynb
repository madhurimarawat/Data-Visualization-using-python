```python
#Experiment 10
#Illustrate Usage of different color codes in python.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating a dataframe
df=pd.DataFrame({'Good 1':[34,17,47,31,20,64,72],'Good 2':[12,29,19,82,65,74,50],'Good 3':[47,78,54,98,96,36,12],'Good 4':[91,63,80,55,78,89,93],'Good 5':[87,83,56,81,11,34,94]},index=["Vendor 1","Vendor 2","Vendor 3","Vendor 4","Vendor 5","Vendor 6","Vendor 7"])
print(df)
```

```python
# Plotting heat map
sns.heatmap(df,annot=True)
plt.title("Heat Map Showing Usage of Color Codes")
plt.xlabel("Goods sold")
plt.ylabel("Vendors")
```

```python
#Another method of Creating this same dataframe
#Importing from excel file
df1=pd.read_excel("/content/Vendors data.xlsx",index_col='Vendors')
print(df1)
```

```python
# Plotting bar graph
df1.plot(kind='bar')
plt.title("Bar Graph Showing Usage of Color Codes")
plt.xlabel("Vendors and Goods sold")
plt.ylabel("Quantity")
```

```python
# Plotting  stacked bar graph
df1.plot(kind='bar',stacked=True)
plt.title("Stacked Bar Graph Showing Usage of Color Codes")
plt.xlabel("Vendors and Goods sold")
plt.ylabel("Quantity")
```

```python
# Plotting scatter plot
sns.scatterplot(df1)
plt.title("Scatter Plot Showing Usage of Color Codes")
plt.xlabel("Vendors and Goods sold")
plt.ylabel("Quantity")
```

```python
# Plotting Heat map
sns.heatmap(df1,annot=True)
plt.title("Heat Map Showing Usage of Color Codes")
plt.xlabel("Goods")
```
