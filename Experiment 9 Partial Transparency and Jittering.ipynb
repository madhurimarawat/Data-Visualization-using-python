```python
#Experiment 9
#Illustrate partial transparency and jittering in python.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Nba.csv")
print(df)
```

```python
sns.scatterplot(data=df,x=df["Height"],y=df["Weight"],hue="Position")
plt.title("Height Vs Weight Graph")
plt.plot(alpha=0.9)
```

```python
sns.swarmplot(data=df,x=df["Height"],y=df["Weight"],hue="Position")
plt.title("Height Vs Weight Graph")
plt.plot(alpha=0.9)
```

```python
#Jittering or Jitter plot
sns.stripplot(data=df,x=df["Height"],y=df["Weight"],hue="Position",jitter=True)
plt.title("Height Vs Weight Graph")
#For partial transparency
#plt.plot(df["Height"], df["Weight"], alpha=0.7)
plt.plot(alpha=0.7)
```
