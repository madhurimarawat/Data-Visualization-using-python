```python
# Experiment 6
# Build a scatter plot and suggest dimension reduction

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
sales_df = pd.read_csv('/content/sales.csv')

# Create a scatterplot of sales vs. units sold
plt.scatter(sales_df['sales'], sales_df['profit'])
plt.ylabel('profit')
plt.title('Sales vs. profit')
plt.show()

# Perform dimension reduction with PCA
pca = PCA(n_components=2)
sales_reduced = pca.fit_transform(sales_df[['sales', 'profit']])

# Create a scatterplot of the reduced data
plt.scatter(sales_reduced[:, 0], sales_reduced[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Reduced Sales Data')
plt.show()
```
