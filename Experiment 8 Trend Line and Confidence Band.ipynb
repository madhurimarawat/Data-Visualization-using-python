```python
# Experiment 8
# Create a trend line with a confidence band in any suitable dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing CSV file
df=pd.read_csv("/content/Nba.csv")
print(df)
```

```python
# Plotting lineplot
sns.lineplot(x=df['Height'],y=df['Weight'])
```

```python
# Plotting Regression Plot
# This dosen't have any relation, this is why the plot looks like this.
sns.regplot(x=df['Salary'],y=df['Weight'],data=df)
```

```python
# Importing another dataset
df1=pd.read_csv("/content/Automobile.csv")
print(df1)
```

```python
# Plotting Regression Plot
sns.regplot(x=df1['length'],y=df1['mileage'],data=df1)
plt.title("Graph showing trend line with a confidence band")
plt.xlabel("Length of Automobile")
plt.ylabel("Mileage of Automobile")
```
